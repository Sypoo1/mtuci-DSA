### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
m = int(input("ВВЕДИТЕ m:"))
n = int(input("ВВЕДИТЕ n:"))
min_limit = int(input("ВВЕДИТЕ min_limit:"))
max_limit = int(input("ВВЕДИТЕ max_limit:"))
matrix = []

for i in range(m):
    row = []
    for j in range(n):
        row.append(random.randint(min_limit,max_limit))
    matrix.append(row)

matrix
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
matrix_choice = copy.deepcopy(matrix)

start_time = time.time()

for ind, row in enumerate(matrix_choice):

    for i in range(len(row) - 1):
        min_value = row[i]
        min_index = i
        for j in range(i + 1, len(row)):
            if row[j] < min_value:
                min_value = row[j]
                min_index = j
                
        if min_index != i:
            row[i], row[min_index] = min_value, row[i]
        

    matrix_choice[ind] = row

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
matrix_choice
```

```python
# Сортировка вставкой.
matrix_insert = copy.deepcopy(matrix)

start_time = time.time()

for ind, row in enumerate(matrix_insert):

    for i in range(1, len(row)):
        for j in range(i, 0, -1):
            if row[j] < row[j-1]:
                row[j], row[j-1] = row[j-1], row[j]
            else:
                break 

    matrix_insert[ind] = row

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
matrix_insert
```

```python
# Сортировка обменом. || Сортировка пузырьком.
matrix_bubble = copy.deepcopy(matrix)

start_time = time.time()

for ind, row in enumerate(matrix_bubble):

    for i in range(len(row) - 1):
        for j in range(len(row) - 1 - i):
            if row[j + 1] < row[j]:
                row[j], row[j + 1] = row[j + 1], row[j]
        

    matrix_bubble[ind] = row

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
matrix_bubble
```

```python
# Сортировка Шелла.
matrix_shell = copy.deepcopy(matrix)
start_time = time.time()

for ind, row in enumerate(matrix_shell):

    gap = len(row) // 2

    while gap > 0:
        for index in range(gap, len(row)):
            current_value = row[index]
            current_index = index

            # сортировка вставкой элементов, с разницой в индексах равной gap
            while current_index >= gap and row[current_index - gap] > current_value:
                row[current_index] = row[current_index - gap]
                current_index -= gap
                row[current_index] = current_value

        gap //= 2
        

    matrix_shell[ind] = row

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
matrix_shell
```

```python
# Быстрая сортировка.
matrix_quick = copy.deepcopy(matrix)
start_time = time.time()

def quick_sort(array: list[int]) -> list[int]:

    if len(array) == 0:
        return array


    based = array[0]

    lower = [x for x in array if x < based]
    center = [x for x in array if x == based]
    higher = [x for x in array if x > based]

    return quick_sort(lower) + center + quick_sort(higher)



for ind, row in enumerate(matrix_quick):

    matrix_quick[ind] = quick_sort(row)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
matrix_quick
    
```

```python
# турнирная сортировка.
matrix_tournament = copy.deepcopy(matrix)
start_time = time.time()

def go(trees):
    # Если список деревьев пуст, возвращаем пустой список
    if not trees:
        return []
    # Выбираем победителя турнира
    winner = play_tournament(trees)
    # Возвращаем список, начинающийся с победителя, и продолжаем с оставшихся деревьев
    return [winner[0]] + go(winner[1:])

def play_tournament(trees):
    # Если в списке только одно дерево, возвращаем его
    if len(trees) == 1:
        return trees[0]
    # Иначе продолжаем турнир с результатами раунда
    return play_tournament(play_round(trees, []))

def play_round(trees, done):
    # Если список деревьев пуст, возвращаем список уже выполненных матчей
    if not trees:
        return done
    # Если в списке только одно дерево, добавляем его в список выполненных матчей
    if len(trees) == 1:
        return trees + done
    # Выбираем победителя матча
    winner = play_game(trees[0], trees[1])
    # Продолжаем раунд с оставшимися деревьями и добавляем победителя в список выполненных матчей
    return play_round(trees[2:], done + [winner])

def play_game(tree1, tree2):
    # Сравниваем корни деревьев и выбираем победителя
    if tree1[0] <= tree2[0]:
        return promote(tree1, tree2)
    else:
        return promote(tree2, tree1)

def promote(winner, loser):
    # Возвращаем новое дерево, где победитель становится корнем, а проигравший добавляется в список детей победителя
    return [winner[0], loser] + winner[1:]

def tournament_sort(alist):
    # Преобразуем входной список в список одноэлементных списков, представляющих деревья
    trees = [[x] for x in alist]
    # Выполняем турнирную сортировку
    return go(trees)


for ind, row in enumerate(matrix_tournament):


    matrix_tournament[ind] = tournament_sort(row)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
matrix_tournament
    
```
